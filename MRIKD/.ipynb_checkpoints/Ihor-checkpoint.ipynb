## Model Przewidywania Zachmurzenia

```python
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_squared_error
```

```python
y = dataset['Średnie dobowe zachmurzenie ogólne [oktanty]']
```

```python
y
```

```python
X = dataset[dataset.columns[1:len(dataset.columns)]]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y.T, test_size=0.2, random_state=12)
```

```python
clf = Ridge(alpha=1.0)
```

```python
clf.fit(X_train, y_train)
```

```python
clf.score(X_test, y_test)
```

```python
from sklearn.linear_model import (
    LinearRegression, TheilSenRegressor, RANSACRegressor, HuberRegressor)
from sklearn.pipeline import make_pipeline
```

```python
estimators = [('OLS', LinearRegression(normalize=True)),
              ('Theil-Sen', TheilSenRegressor(random_state=42)),
              ('RANSAC', RANSACRegressor(random_state=42)),
              ('HuberRegressor', HuberRegressor()),
              ('Ridge', Ridge())]
```

```python
for name, estimator in estimators:
        model = make_pipeline(estimator)
        model.fit(X_train, y_train)
        mse = mean_squared_error(model.predict(X_test), y_test)
        print(model.score(X_test, y_test))
        print(mse)
        print(name + " Done")
```

```python
clasifiers = [
    ('AdaBoost', AdaBoostClassifier(n_estimators=100, random_state=0)),
    ('SVM', SVC(gamma='auto'))
]
```
